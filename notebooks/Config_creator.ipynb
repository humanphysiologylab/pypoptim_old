# Voigt template

```python
import copy
```

```python
# filename_template = "../configs/maleckar/config_voigt_ical_ghk_rel_template.json"
# filename_template = "../configs/maleckar/config_rheeda_template.json"

# filename_template = "../configs/koivumaki/config_voigt_template.json"
# filename_template = "../configs/koivumaki/config_voigt_fluo_template.json"
filename_template = "../configs/koivumaki/config_voigt_rel_template.json"

# filename_template = "../configs/grandi/config_voigt_fluo_template.json"


with open(filename_template) as f:
    text = f.read()
    text = strip_comments(text)
    config_template = json.loads(text)
```

```python
ec_template = config_template['experimental_conditions']['CL_VALUE']
ec_template
```

```python
group_indices = [1, 2, 3, 4]
cell_indices = [3, 4, 5, 8]

CL_list_1 = [2064, 1032, 516, 344, 258]
CL_list_2 = [516, 344, 258, 206, 172] # G2 C1 & C2

dirname_pypoptim_abs = os.path.abspath("../../pypoptim")
dirname_config_abs = os.path.abspath("../configs/koivumaki/rel")

os.makedirs(dirname_config_abs, exist_ok=True)

for ig in group_indices:
    for ic in range(1, cell_indices[ig-1] + 1):
        
        filename = os.path.join(dirname_config_abs, f"config_G{ig}C{ic}.json")
        
        config = copy.deepcopy(config_template)
        
        CL_list = CL_list_2 if ig == 2 and ic <= 2 else CL_list_1
        
        for CL in CL_list:
            ec = copy.deepcopy(ec_template)
            stim_period_legend_name = config_template['stim_period_legend_name']
            ec['params'][stim_period_legend_name] = CL / 1000 if 'grandi' not in filename_template else CL
            ec['filename_phenotype'] = ec['filename_phenotype'].replace('CL_PHENOTYPE', str(CL))

            for CL_STATE in 2000, 1000, 666, 500:
                if CL >= CL_STATE:
                    break
            ec['filename_state'] = ec['filename_state'].replace('CL_STATE', str(CL_STATE))
            
            config['experimental_conditions'][str(CL)] = ec

        del config['experimental_conditions']['CL_VALUE']     
        
        for key, value in config.items():
            if 'filename' in key or 'folder_name' in key:
                path_value_abs = os.path.abspath(os.path.join(dirname_pypoptim_abs, value))
                path_value_rel = os.path.relpath(path_value_abs, dirname_config_abs)
                config[key] = path_value_rel
                
            if isinstance(value, dict):
                for key2, value2 in value.items():
                    if 'filename' in key2 or 'folder_name' in key2:
                        path_value_abs = os.path.abspath(os.path.join(dirname_pypoptim_abs, value2))
                        path_value_rel = os.path.relpath(path_value_abs, dirname_config_abs)
                        config[key][key2] = path_value_rel
                        
                    if isinstance(value2, dict):
                        for key3, value3 in value2.items():
                            if 'filename' in key3 or 'folder_name' in key3:
                                path_value_abs = os.path.abspath(os.path.join(dirname_pypoptim_abs, value3))
                                path_value_rel = os.path.relpath(path_value_abs, dirname_config_abs)
                                config[key][key2][key3] = path_value_rel
                    
        with open(filename, 'w') as f:
            text = json.dumps(config, indent=2)
            text = text.replace('GROUPINDEX', str(ig))
            text = text.replace('CELLINDEX', str(ic))
            f.write(text)
```
